10) Prime Number Checker
Ask the user for a number and check if it's a prime number.

```python
n = input("input the number")
n = int(n)
if n <= 1:
    print(False)
else:
    is_prime = True
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            is_prime = False
            break
    print(is_prime)
```

13- Write a program that remove all vowels from the input word and
generate a brief version of it

---

```python
def v(word):
    vowels = "aeiouAEIOU"
    brief_version = ""
    
    for char in word:
        if char not in vowels:
            brief_version += char
            
    return brief_version

user_word = input("Enter a word to abbreviate: ")
print(f"The brief version is: {v(user_word)}")
```

14- Write a program that counts up the number of vowels **[a, e, i, o,u]** contained in the string.


```python
def v(word):
    vowels = "aeiouAEIOU"
    c = 0
    
    for char in word:
        if char in vowels:
            c+= 1
            
    return c

user_word = input("Enter a word to abbreviate: ")
print(f"The vowels count is : {v(user_word)}")
```
