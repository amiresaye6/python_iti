--  example: madam

```python
str = input("input a string")
str = str.upper()
c = 0
print(f"Upper case string\n {str}")
for ch in str:
    if ch in "AEIOU":
        c+= 1

print(f"number of vouls is {c}")

str = str[::-1]

print(f"reversed string \n {str}")
f= True
for i in range(len(str) // 2 + 1):
    if str[i] != str[len(str) - i - 1]:
        f = False
        break
if f:
    print("string is palendrome :__:")
else:
    print("string not palindrome")
```

### 3- Set-Based Password Security Checker

## Task:
Implement a function that checks if a password is weak by comparing it to a predefined set of common passwords.

**Common passwords:**
- `"123456"`, `"password"`, `"123456789"`, `"12345"`, `"12345678"`, `"qwerty"`
- `"1234567"`, `"111111"`, `"123123"`, `"abc123"`, `"password1"`, `"1234"`

## **Task 1: Working with Tuples**

- Print the first and last fruit.
- Slice the tuple to get the first three fruits.
- Count the occurrences of a specific fruit.
- Try modifying the tuple and handle the error gracefully.

---

```python
fruits = ("apple", "banana", "kiwi", 'apple', "dragon fruit", "watermelon")

print(f"first fruit {fruits[0]}, last fruit {fruits[-1]}")
print(f"first 3 fruit {fruits[:3]}")
fruits.count('apple')

try:
    fruits[0] = "banana" # will raise a typeerror, tuples ar immutable, we can not change or udpate after creation
except TypeError:
    print("tuples are immutable")

```
