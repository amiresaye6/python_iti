### **1. squares numers:**
- Create a dictionary where keys are the first 5 natural numbers and values are their squares.
- Write a script to:
  - Print all keys and their respective values.
  - Add a new key-value pair: `6: 36`.
  - Check if a specific key (e.g., 4) exists in the dictionary.


```python
vals = {
    "2": 4,
    "3": 9,
    "5": 25,
    "7": 49,
    "11": 121
}

print("before")
for (k, v) in vals.items():
    print(f'key: {k}:, val: {v}')

vals["6"] = 36

print("aftere")
for (k, v) in vals.items():
    print(f'key: {k}:, val: {v}')

if ("4" in vals):
    print('4 is present')
else: 
    print("no 4 in vals")

```

### **2. Student Grades:**
- Create a dictionary to store student names as keys and a list of their grades as values.
- Write a script to:
  - Calculate the average grade for each student.
  - Find the student with the highest average.

```python
grades = {
    "Amir Alsayed": [100, 100, 100, 100, 100],
    "Amir Maula": [99, 99, 100, 100, 100],
    "Hamdy": [100, 100, 100, 100, 98]
}

highestAvg = 0
stdWithHightestGrade = ""

for name, g in grades.items():
    avgGrade = sum(g) / len(g)
    if avgGrade >= highestAvg:
        highestAvg = avgGrade
        stdWithHightestGrade = name
    print(f'student {name} has avrage grade of {avgGrade}%')

print(f"student with the hightest grad is {stdWithHightestGrade} with avg grade of {highestAvg} %")
```

### **3. Shopping Cart:**

### **4. Employee Database:**
- Create a list of dictionaries where each dictionary represents an employee with attributes:
  - `name`, `age`, and `department`.
- Write a script to:
  - Filter employees by a specific department.
  - Find the average age of employees.

```python
employees = [
    {
        "name": "Amir Alsayed", 
        "age": 23,
        "department": "Backend"
    },
    {
        "name": "Amir Maula", 
        "age": 23,
        "department": "Backend"
    },
    {
        "name": "Mohamed Hamdy", 
        "age": 22,
        "department": "Frontend"
    },
    {
        "name": "Mostafa Assad", 
        "age": 24,
        "department": "DevOps"
    },
    {
        "name": "some other old man", 
        "age": 50,
        "department": "Manegerial"
    },
]

def filterEmp(dep, employees):
    res = []
    for emp in employees:
        if emp["department"] == dep:
            res.append(emp)
    return res
avgAge = 0
for em in employees:
    avgAge += em["age"]

print(f"avrage age of all employees is: {avgAge / len(employees)} years")

filterEmp("Backend", employees)
```
